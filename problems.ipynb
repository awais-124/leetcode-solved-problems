Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]
 

Constraints:

2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109
Only one valid answer exists.

```python
class Solution:
    def twoSum(self, nums, target):
        for i in range(0,len(nums)):
            for j in range(i+1,len(nums)):
                if target - nums[i] == nums[j]:
                    return [i,j]
```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        char_map = {}
        left = 0
        max_length = 0

        for right in range(len(s)):
            if s[right] in char_map and char_map[s[right]] >= left:
                left = char_map[s[right]] + 1

            char_map[s[right]] = right
            max_length = max(max_length, right - left + 1)

        return max_length
```

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        def expandAroundCenter(s, left, right):
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right += 1
            return s[left + 1:right]

        if not s or len(s) == 1:
            return s
    
        longest_palindrome = ""
        
        for i in range(len(s)):
            # Odd length palindromes (centered at one character)
            palindrome1 = expandAroundCenter(s, i, i)
            # Even length palindromes (centered between two characters)
            palindrome2 = expandAroundCenter(s, i, i + 1)
            
            # Update the longest palindrome found so far
            longest_palindrome = max(longest_palindrome, palindrome1,    palindrome2, key=len)
        
        return longest_palindrome
        
```

```python
class Solution:
    def convert(self, s: str, numRows: int) -> str:
        if numRows == 1 or numRows >= len(s):
            return s

        # Create an array for each row
        rows = [''] * numRows
        cur_row = 0
        going_down = False

        # Traverse the string
        for char in s:
            rows[cur_row] += char  # Add character to the current row
            # Change direction when you hit the top or bottom row
            if cur_row == 0 or cur_row == numRows - 1:
                going_down = not going_down
            # Move to the next row, either down or up
            cur_row += 1 if going_down else -1

        # Join all rows to form the final result
        return ''.join(rows)
```

```python
class Solution:
    def myAtoi(self, s: str) -> int:
        # Initialize variables
        i = 0
        n = len(s)
        result = 0
        sign = 1
        INT_MAX = 2**31 - 1
        INT_MIN = -2**31
        
        # Step 1: Ignore leading whitespace
        while i < n and s[i] == ' ':
            i += 1
        
        # Step 2: Check for a sign
        if i < n and (s[i] == '-' or s[i] == '+'):
            sign = -1 if s[i] == '-' else 1
            i += 1
        
        # Step 3: Convert digits to integer
        while i < n and s[i].isdigit():
            digit = int(s[i])
            
            # Step 4: Check for overflow/underflow
            if result > (INT_MAX - digit) // 10:
                return INT_MIN if sign == -1 else INT_MAX
            
            result = result * 10 + digit
            i += 1
        
        # Apply sign and return the result
        return result * sign
```

```python
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        # Initialize a DP table where dp[i][j] is True if s[:i] matches p[:j]
        dp = [[False] * (len(p) + 1) for _ in range(len(s) + 1)]
        
        # Empty string and empty pattern match
        dp[0][0] = True
        
        # Fill the DP table for cases where p has '*' patterns
        for j in range(1, len(p) + 1):
            if p[j - 1] == '*':
                dp[0][j] = dp[0][j - 2]  # '*' can eliminate the preceding element
        
        # Fill the DP table
        for i in range(1, len(s) + 1):
            for j in range(1, len(p) + 1):
                # Case 1: If characters match, or if p[j-1] is '.', we carry over the previous match
                if p[j - 1] == s[i - 1] or p[j - 1] == '.':
                    dp[i][j] = dp[i - 1][j - 1]
                
                # Case 2: If p[j-1] is '*', check if we can match zero or more of the preceding character
                elif p[j - 1] == '*':
                    # Check zero occurrence (skip the previous character and '*')
                    dp[i][j] = dp[i][j - 2]
                    
                    # Check one or more occurrence of the character before '*'
                    if p[j - 2] == s[i - 1] or p[j - 2] == '.':
                        dp[i][j] = dp[i][j] or dp[i - 1][j]
        
        # The result is in dp[len(s)][len(p)]
        return dp[len(s)][len(p)]
```

```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        # Mapping of digits to corresponding letters
        digit_to_char = {
            '2': 'abc', '3': 'def', '4': 'ghi',
            '5': 'jkl', '6': 'mno', '7': 'pqrs',
            '8': 'tuv', '9': 'wxyz'
        }

        # Result list to store all combinations
        result = []
        
        # If the input digits are empty, return an empty list
        if not digits:
            return result
        
        # Helper function to perform backtracking
        def backtrack(index: int, current_combination: str):
            # If the current combination is the same length as digits, add to result
            if index == len(digits):
                result.append(current_combination)
                return
            
            # Get the characters mapped to the current digit
            letters = digit_to_char[digits[index]]
            
            # Iterate over each letter and proceed to the next digit
            for letter in letters:
                backtrack(index + 1, current_combination + letter)
        
        # Start backtracking from the first digit
        backtrack(0, "")
        
        return result
```

```python
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        # Sort the array first
        nums.sort()
        result = []
        n = len(nums)
        
        # Iterate over the array with two nested loops
        for i in range(n - 3):
            # Skip duplicate values for the first number
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            for j in range(i + 1, n - 2):
                # Skip duplicate values for the second number
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue
                
                # Now apply the two-pointer approach
                left, right = j + 1, n - 1
                while left < right:
                    total_sum = nums[i] + nums[j] + nums[left] + nums[right]
                    
                    if total_sum == target:
                        # Found a valid quadruplet
                        result.append([nums[i], nums[j], nums[left], nums[right]])
                        
                        # Move left pointer forward and avoid duplicates
                        while left < right and nums[left] == nums[left + 1]:
                            left += 1
                        left += 1
                        
                        # Move right pointer backward and avoid duplicates
                        while left < right and nums[right] == nums[right - 1]:
                            right -= 1
                        right -= 1
                    
                    # If sum is too small, move the left pointer to increase the sum
                    elif total_sum < target:
                        left += 1
                    # If sum is too large, move the right pointer to decrease the sum
                    else:
                        right -= 1
        
        return result
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # Create a dummy node to simplify edge cases (like removing the head)
        dummy = ListNode(0)
        dummy.next = head
        
        # Initialize two pointers: fast and slow
        fast = slow = dummy
        
        # Move fast pointer n+1 steps ahead (so slow ends up just before the target node)
        for _ in range(n + 1):
            fast = fast.next
        
        # Move both pointers until fast reaches the end of the list
        while fast:
            fast = fast.next
            slow = slow.next
        
        # Now, slow is just before the node to be removed
        # Skip the nth node
        slow.next = slow.next.next
        
        # Return the new head (dummy.next)
        return dummy.next
```

```python
class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        
        # Iterate through the grid and update the path sums
        for i in range(m):
            for j in range(n):
                if i == 0 and j == 0:
                    continue  # Skip the top-left corner (starting point)
                elif i == 0:
                    grid[i][j] += grid[i][j - 1]  # Only move from the left
                elif j == 0:
                    grid[i][j] += grid[i - 1][j]  # Only move from the top
                else:
                    grid[i][j] += min(grid[i - 1][j], grid[i][j - 1])  # Min of top and left
        
        # The bottom-right cell contains the minimum path sum
        return grid[-1][-1]
```

```python
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        m, n = len(matrix), len(matrix[0])
        is_col = False  # To track if the first column needs to be zeroed
        
        # Mark rows and columns to be zeroed
        for i in range(m):
            if matrix[i][0] == 0:
                is_col = True
            for j in range(1, n):
                if matrix[i][j] == 0:
                    matrix[i][0] = 0
                    matrix[0][j] = 0
        
        # Update the matrix using markers
        for i in range(1, m):
            for j in range(1, n):
                if matrix[i][0] == 0 or matrix[0][j] == 0:
                    matrix[i][j] = 0
        
        # Handle the first row
        if matrix[0][0] == 0:
            for j in range(n):
                matrix[0][j] = 0
        
        # Handle the first column
        if is_col:
            for i in range(m):
                matrix[i][0] = 0
```

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        if not matrix or not matrix[0]:
            return False
        
        m, n = len(matrix), len(matrix[0])
        left, right = 0, m * n - 1
        
        while left <= right:
            mid = (left + right) // 2
            # Convert 1D index to 2D coordinates
            row, col = divmod(mid, n)
            mid_val = matrix[row][col]
            
            if mid_val == target:
                return True
            elif mid_val < target:
                left = mid + 1
            else:
                right = mid - 1
        
        return False
```

```python
from collections import deque, defaultdict

class Solution:
    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        # If endWord is not in wordList, no solution exists
        if endWord not in wordList:
            return 0
        
        # Create a dictionary of word patterns
        word_patterns = defaultdict(list)
        word_length = len(beginWord)
        
        # Create patterns for each word in wordList and beginWord
        for word in wordList + [beginWord]:
            for i in range(word_length):
                pattern = word[:i] + '*' + word[i+1:]
                word_patterns[pattern].append(word)
        
        # Queue for BFS
        queue = deque([(beginWord, 1)])
        visited = set([beginWord])
        
        # BFS
        while queue:
            current_word, level = queue.popleft()
            
            # Check all possible one-letter transformations
            for i in range(word_length):
                pattern = current_word[:i] + '*' + current_word[i+1:]
                
                # Check all words matching this pattern
                for word in word_patterns[pattern]:
                    if word == endWord:
                        return level + 1
                    
                    if word not in visited:
                        visited.add(word)
                        queue.append((word, level + 1))
        
        # If we've exhausted all possibilities without finding endWord
        return 0
```
